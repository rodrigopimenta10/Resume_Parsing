```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os

r = requests.get('http://www.4icu.org/cn/')

soup = BeautifulSoup(r.content, "lxml")
```

```python
print (soup)
```

```python
from IPython.display import HTML
type (soup)

letters = soup.find_all(class_="lead")

print (letters)
```

```python
university = {}
for element in letters:
    university[element.get_text()] = {}
    
print (university)
```

```python
type(university)
```

```python
university.keys()
```

```python
list_keys = [ k for k in university.keys() ]

print(list_keys)
```

```python
df = pd.DataFrame(list_keys)
print (df)
```

```python
writer = pd.ExcelWriter('China_University.xlsx')
df.to_excel(writer, index = False)
writer.save()
```
